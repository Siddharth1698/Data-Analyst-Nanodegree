```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("results.csv")
```

```python
df 
```

```python
df.set_index('year')['global_avg'].plot(figsize=(12, 10), linewidth=2, color='maroon')
plt.title('Line chart of global data')
plt.ylabel('Global Average Temp')
plt.xlabel('Year')
```

```python
df_ma = df.rolling(window=20).mean()
```

```python
df_ma.set_index('year')['global_avg'].plot(figsize=(12, 10), linewidth=2, color='maroon')
plt.title('Line chart of global data over Moving Average of 20 years')
plt.ylabel('Global Average Temp')
plt.xlabel('Year')
```

```python
plt.figure()

x = df_ma['year']
y1 = df_ma['global_avg']
y2 = df_ma['city_avg']

plt.plot(x,y1)
plt.plot(x,y2)
plt.title('Line chart of Banglore(Orange) and World(Blue) averages')
plt.xlabel('Year')
plt.ylabel('Average Temp')
```
